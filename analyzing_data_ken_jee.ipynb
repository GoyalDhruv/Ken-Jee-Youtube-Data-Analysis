```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Aggregated Metrics By Video

```python
video=pd.read_csv('Aggregated_Metrics_By_Video.csv')
```

```python
video
```

We can see that first row is summarized as `Total`

```python
total_video=video[:1]
```

We will drop the first row.

```python
video.drop(video.index[:1],inplace=True)
```

```python
video.shape
```

```python
video.info()
```

```python
video.isnull().sum()
```

There are 2 missing vlaues in CRM variable.

```python
video[video['CPM (USD)'].isnull()]
```

We will replace it by 0.

```python
video['CPM (USD)']=video['CPM (USD)'].fillna(0)
```

We will convert Video publish time feature to appropriate Date data type from character

We will fix the date from Video Publish time and break it also down into new features of Year and Month

```python
video['Video pub­lish time'] = pd.to_datetime(video['Video pub­lish time'], format = "%b %d, %Y")
video['Video pub­lish time']
```

```python
video['Year'] = video['Video pub­lish time'].dt.year
video['Month'] = video['Video pub­lish time'].dt.month
```

Number Of Published Videos By Year

```python
video['Year'].value_counts().sort_index()
```

```python
count=video['Year'].value_counts().sort_index()
plt.bar(count.index, count.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number Of Published Videos By Year')
```

Estimated Revenue By Year

```python
video.groupby(['Year'])['Your es­tim­ated rev­en­ue (USD)'].sum()
```

```python
revenue=video.groupby(['Year'])['Your es­tim­ated rev­en­ue (USD)'].sum()
plt.bar(revenue.index, revenue.values)
plt.title('Estimated Revenue By Year')
```

# Video Performance Over Time

```python
performance=pd.read_csv('Video_Performance_Over_Time.csv')
```

```python
performance
```

```python
performance.shape
```

```python
performance.info()
```

```python
performance.isnull().sum()
```

Converting Date Dtype from object to date

```python
performance['Date'] = performance['Date'].str.replace('Sept','Sep')
```

```python
performance['Date'] = pd.to_datetime(performance['Date'], format = "%d %b %Y")
```

Views per Date

```python
performance['Year'] = performance['Date'].dt.year
performance['Month'] = performance['Date'].dt.month
```

```python
plt.plot(performance['Date'],performance['Views'])
plt.xticks(rotation=45)
plt.title('Views per Date')
plt.show()
```

Subscribers added by Date

```python
plt.plot(performance['Date'],performance['User Subscriptions Added'])
plt.xticks(rotation=45)
plt.title('Subscribers per Date')
plt.show()
```

# Comments

```python
comments=pd.read_csv('All_Comments_Final.csv')
```

```python
comments
```

```python
comments.shape
```

```python
comments.info()
```

```python
comments.isnull().sum()
```

```python
comments['Comments'] = comments['Comments'].astype(str)
```

```python
comments['Date']=pd.to_datetime(comments['Date'])
```

```python
comments['Year'] = comments['Date'].dt.year
comments['Month'] = comments['Date'].dt.month
comments['Day'] = comments['Date'].dt.day
```

Comments by Year

```python
comments.groupby('Year')['Comments'].count()
```

```python
comment_year=comments.groupby('Year')['Comments'].count()
plt.bar(comment_year.index,comment_year.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Comments by Year')
plt.show()
```

Most Liked Comment

```python
comments.groupby('Comments')['Like_Count'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


text = ' '.join(comment for comment in comments['Comments'])


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
